Q1. What is the difference between a function and a method in Python?
 -
 A function in Python is a standalone block of reusable code defined using the def keyword, not tied to any object. A method, on the other hand, is a function defined within a class and is associated with an object. Methods typically operate on the object's data and require self as the first parameter (for instance methods). Functions are called directly, while methods are called on objects.
 In Python, functions and methods are similar in that both are callable objects, but they differ in how they are associated with objects and how they are used. Here's the key difference



1. Function
A function is a block of code that performs a specific task.

It is not associated with any particular object and can be called independently.

Functions are defined using the def keyword and are often part of the global scope or a module.

2. Method
A method is a function that is associated with an object (it is a member of a class).

It is called on an instance of a class, and it can access and modify the attributes of the object.

Methods are defined within a class and typically take self as their first parameter (for instance methods).

Q6. Explain the concept of generators in Python and how they are defined?
- Generators in Python are a type of iterable, like lists or tuples, but they are lazy evaluated, meaning they generate items one at a time as you iterate over them, instead of storing them in memory all at once. This makes them more memory efficient for handling large data sets.

How are generators defined?
Generators are defined using either:

A generator function with the yield keyword.
A generator expression, which is similar to a list comprehension.

##Key Points:

yield: Pauses the function, saving its state, and returns a value. The function can resume from where it left off when called again.
Memory Efficient: Generators do not store all values in memory; they produce values on demand.

Q7, What are the advantages of using generators over regular functions?
- The advantages of using generators over regular functions are:

Memory Efficiency:

Generators yield values one at a time and do not store the entire sequence in memory, making them more memory-efficient, especially for large datasets.

Lazy Evaluation:

 Generators produce values only when needed, which can improve performance by avoiding unnecessary computations.

Improved Performance:

 For large or infinite sequences, generators avoid the overhead of building and holding large data structures, leading to faster execution.

State Retention:

 Generators automatically save their state between yields, so they can continue where they left off without needing to maintain complex state management manually.

Overall, generators allow for more efficient and scalable handling of large or infinite data streams.

Q10, What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

- The map(), reduce(), and filter() functions in Python are all higher-order functions used for applying operations on iterables, but they differ in their purposes and behavior:

###1. map():
Purpose: Applies a given function to every item of an iterable and returns a map object (iterator) of the results.

Usage: When you need to transform each item in an iterable.

### 2. reduce():
Purpose: Applies a function of two arguments cumulatively to the items of an iterable, reducing the iterable to a single value. It is available in the functools module.

Usage: When you need to perform a rolling computation on the iterable.

###3. filter():
Purpose: Applies a function that returns a boolean value (True or False) to each item in an iterable, filtering out the elements for which the function returns False.

Usage: When you need to filter items in an iterable based on some condition.

###Key Differences:


map(): Transforms each item in the iterable (returns a new iterable of transformed values).
reduce(): Reduces the iterable to a single cumulative result (returns a single value).
filter(): Filters elements based on a condition (returns an iterable of elements that satisfy the condition).

Q1, Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list?

```python

#Here's a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list and Explanation:
#sum_of_even_numbers is the function name.
#numbers is the input list of numbers.
#The function uses a generator expression num for num in numbers if num % 2 == 0 to iterate through the list and filter only the even numbers.
#sum() computes the sum of the filtered numbers.


def sum_of_even_numbers(numbers):
    return sum(num for num in numbers if num % 2 == 0)

# Example usage

numbers = [1, 2, 3, 4, 5, 6]
result = sum_of_even_numbers(numbers)
print(result)  # Output: 12



```

Q4, Write a Python function that checks if a given number is prime or not from 1 to 200?

```python
#Here’s a Python function that checks if a given number is prime or not, specifically for numbers between 1 and 200
#Explanation:
#The function is_prime(n) checks if a number n is prime.
#If n <= 1, it immediately returns False because numbers less than or equal to 1 are not prime.
#It iterates through potential divisors from 2 to the square root of n (rounded up). If n is divisible by any number in this range, it returns False (meaning it's not prime).
#If no divisors are found, the function returns True, indicating the number is prime.
#The for loop from 1 to 200 prints the prime numbers in that range.

def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

# Example usage
for number in range(1, 201):
    if is_prime(number):
        print(number, "is prime")
```

Q5, Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms?

```python
#Here’s a Python iterator class that generates the Fibonacci sequence up to a specified number of terms
#Explanation:
#__init__(self, terms): This initializes the iterator with the number of terms (terms) in the Fibonacci sequence to generate. It also initializes the first two Fibonacci numbers (a and b) and a counter (count).
#__iter__(self): This method makes the class iterable by returning the iterator object itself.
#__next__(self): This method generates the next number in the Fibonacci sequence and raises StopIteration when the sequence has generated the specified number of terms (terms).

class FibonacciIterator:
    def __init__(self, terms):
        self.terms = terms
        self.a, self.b = 0, 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count < self.terms:
            result = self.a
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return result
        else:
            raise StopIteration

# Example usage
fibonacci = FibonacciIterator(10)
for num in fibonacci:
    print(num)
```

Q6, Write a generator function in Python that yields the powers of 2 up to a given exponent?

```python
#Here's a Python generator function that yields the powers of 2 up to a given exponent:
"""
Explanation:
powers_of_2(exponent): This is the generator function that takes an exponent as input and yields the powers of 2 from
2
0
2
0
  up to
2
exponent
2
exponent
 .
The for loop iterates from 0 to the given exponent (inclusive), and on each iteration, yield 2 ** i yields the value of
2
𝑖
2
i
 .
You can use this generator in a loop, which will call __next__() automatically until the specified exponent is reached.
"""

def powers_of_2(exponent):
    for i in range(exponent + 1):
        yield 2 ** i

# Example usage
for power in powers_of_2(5):
    print(power)
```

Q7, Implement a generator function that reads a file line by line and yields each line as a string?

```python
# Here’s a Python generator function that reads a file line by line and yields each line as a string:
#Explanation:
#read_file_line_by_line(file_path): This is the generator function that takes the file_path as input.
#The with open(file_path, 'r') as file: statement opens the file in read mode and automatically handles closing the file after reading.
#The for line in file: loop iterates through the file line by line.
#The yield line.strip() yields each line of the file after stripping the newline character (to avoid extra blank lines when printing).

def read_file_line_by_line(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line.strip()  # Use strip() to remove the newline character



```

Q9, Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit?

```python
#Here’s a Python program that uses the map() function to convert a list of temperatures from Celsius to Fahrenheit:

#Explanation:
#The list celsius_temperatures contains the temperatures in Celsius.
#The function celsius_to_fahrenheit(celsius) converts each Celsius value to Fahrenheit using the formula:
# 𝐹=(𝐶×95)+32F=(C×59​ )+32
#The map() function applies celsius_to_fahrenheit to each element in celsius_temperatures and returns a map object. The list() function is used to convert it to a list.
#The result is stored in fahrenheit_temperatures and printed.

# List of temperatures in Celsius
celsius_temperatures = [0, 10, 20, 30, 40]

# Function to convert Celsius to Fahrenheit
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

# Use map() to apply the conversion function to the list
fahrenheit_temperatures = list(map(celsius_to_fahrenheit, celsius_temperatures))

# Example usage
print(fahrenheit_temperatures)
```

10, Create a Python program that uses `filter()` to remove all the vowels from a given string?

```python
#Here’s a Python program that uses the filter() function to remove all vowels from a given string
#Explanation:
#is_not_vowel(char): This function checks if a character is not a vowel by checking if it is not in the string 'aeiou'.
#filter(is_not_vowel, input_string): The filter() function applies is_not_vowel to each character of input_string. It returns an iterator with only the characters that are not vowels.
#''.join(): Since filter() returns an iterator, we use ''.join() to combine the characters back into a string.


# Function to check if a character is a vowel
def is_not_vowel(char):
    return char.lower() not in 'aeiou'

# Input string
input_string = "Hello, World!"

# Use filter() to remove vowels
filtered_string = ''.join(filter(is_not_vowel, input_string))

# Example usage
print(filtered_string)
```

Q11, Imagine an accounting routine used in a book shop. It works on a list with sublists, which look like this?
Write a Python program, which returns a list with 2-tuples. Each tuple consists of the order number and the
product of the price per item and the quantity. The product should be increased by 10,- € if the value of the
order is smaller than 100,00 €.

Write a Python program using lambda and map

```python
#To solve this problem, we can use Python's map() function along with lambda to transform the input list into the desired output. Here's a Python program that processes an order list as described:

#Problem Breakdown:
#The list consists of sublists where each sublist contains the order number, the product price per item, and the quantity.
#We need to calculate the total price per order (price * quantity).
#If the total price is less than 100.00€, we need to increase the product by 10€.
#The result should be returned as a list of 2-tuples, each containing the order number and the final price.

#Explanation:
#Input: The list orders contains sublists, each with three elements: order number, price per item, and quantity.
#Lambda function: The lambda function calculates the total price (order[1] * order[2]) and adds 10€ if the total price is less than 100. If the total price is 100€ or more, no addition is made.
#Map: The map() function applies the lambda function to each order in the orders list.
#Result: The result is converted into a list of tuples, where each tuple contains the order number and the final price.


# Sample input list: [order_number, price_per_item, quantity]
orders = [
    [1, 12.50, 5],  # Order 1
    [2, 20.00, 3],  # Order 2
    [3, 45.00, 1],  # Order 3
    [4, 10.00, 8]   # Order 4
]

# Using map and lambda to process each order and compute the final price
result = list(map(lambda order: (order[0], order[1] * order[2] + 10 if order[1] * order[2] < 100 else order[1] * order[2]), orders))

# Output the result
print(result)
```
